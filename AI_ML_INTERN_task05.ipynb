```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score

# Define a professional sample dataset for support tickets
training_tickets = [
    "My laptop screen is flickering and not displaying properly.",
    "Cannot connect to the office Wi-Fi network from my phone.",
    "The ERP software keeps crashing whenever I generate a report.",
    "The printer is jammed and refusing to print documents.",
    "VPN connection fails with a timeout error on my remote desktop.",
    "Need access to the SQL database server for the new project.",
    "Keyboard keys are not working after the coffee spill.",
    "The CRM application is showing a 500 internal server error."
]

# Labels: 0 = Hardware, 1 = Network, 2 = Software
training_labels = [0, 1, 2, 0, 1, 1, 0, 2]

# Test dataset to evaluate the performance
testing_tickets = [
    "The mouse click is unresponsive and lagging.",
    "Internet is down and websites are loading extremely slow.",
    "Excel sheets are freezing when launching the plugin."
]
testing_labels = [0, 1, 2]

# Define class mappings for professional presentation
class_names = ["Hardware Issue", "Network Issue", "Software Issue"]

print("Cell 1: Custom support ticket dataset initialized successfully.")
```

```python
# Create a machine learning pipeline combining text vectorization and classification
ticket_classifier_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english')),
    ('classifier', LogisticRegression(C=1.0, max_iter=1000))
])

# Train the text classification model
ticket_classifier_pipeline.fit(training_tickets, training_labels)

print("Cell 2: Text classification pipeline trained successfully.")
```

```python
# Generate predictions on the test dataset
predicted_labels = ticket_classifier_pipeline.predict(testing_tickets)

# Output evaluation metrics
print("\n--- Support Ticket Classification Metrics ---")
print(f"Accuracy Score: {accuracy_score(testing_labels, predicted_labels):.4f}")
print("\nClassification Report:\n")
print(classification_report(testing_labels, predicted_labels, target_names=class_names))

# Run a quick interactive test sample
sample_query = ["Our team cannot access the main server router."]
sample_pred = ticket_classifier_pipeline.predict(sample_query)
print(f"Sample Query: '{sample_query[0]}'")
print(f"Predicted Category: {class_names[sample_pred[0]]}")
```

```python
import joblib

# Save the trained text classification pipeline to disk
joblib.dump(ticket_classifier_pipeline, 'support_ticket_classifier.pkl')
print("Model pipeline saved successfully using joblib.")
```
